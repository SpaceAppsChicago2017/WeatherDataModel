## Merge the features file and the prediction file

```python
import pandas as pd

# load files
feature_file = pd.read_csv('feature_data_2.csv')
prediction_file = pd.read_csv('selected.csv')
```

```python
prediction_file['day'] = [int(str(x)[6:]) for x in prediction_file.date]
prediction_file['month'] = [int(str(x)[4:6]) for x in prediction_file.date]
prediction_file['year'] = [int(str(x)[:4]) for x in prediction_file.date]
prediction_file['merge_key'] = ["%d-%d-%d-%s-%s"%(x[1].year, x[1].month, x[1].day,
                                                  x[1].county, x[1].state)
                                for x in prediction_file.iterrows()]
```

```python
print prediction_file
print len(set(list(prediction_file['merge_key'])))
```

```python
print list(feature_file)
feature_file['day'] = [int(x.split('/')[1]) for x in feature_file['Date.Local']]
feature_file['month'] = [int(x.split('/')[0]) for x in feature_file['Date.Local']]
feature_file['year'] = [int(x.split('/')[2]) for x in feature_file['Date.Local']]
#feature_file['ncounty'] = [x for x in feature_file['County.Name']]
#feature_file['nstate'] = [x for x in feature_file['State.Name']]
feature_file['merge_key'] = ["20%d-%d-%d-%s-%s"%(x[1]['year'], x[1]['month'], x[1]['day'],
                                        x[1]['County.Name'], x[1]['State.Name'])
                                for x in feature_file.iterrows()]
```

```python
print feature_file
```

```python
# select feature entries in the geo square
min_lat = 39.66
max_lat = 45.41
min_lng = -93.64
max_lng = -86.00

selected_features = feature_file.loc[(min_lat < feature_file.Latitude) &
                                     (feature_file.Latitude < max_lat) &
                                     (min_lng < feature_file.Longitude) &
                                     (feature_file.Longitude < max_lng)]
```

```python
merged_data = pd.merge(selected_features, prediction_file, on = 'merge_key', how = 'left')
print merged_data
```

```python
print list(merged_data)
```

```python
merged_data = merged_data[['merge_key', 'day_x', 'month_x', 'year_x',
                           'County.Name', 'State.Name', 'Latitude', 'Longitude', 
                          'Wind_Arithmetic.Mean', 'Wind_X1st.Max.Value', 'Wind_X1st.Max.Hour',
                          'temp_Arithmetic.Mean', 'temp_X1st.Max.Value', 'temp_X1st.Max.Hour',
                          'pressure_Arithmetic.Mean', 'pressure_X1st.Max.Value', 'pressure_X1st.Max.Hour',
                          'type']]
```

```python
merged_data['type'] = merged_data['type'].fillna('Normal')
merged_data.to_csv('merged_feature_and_prediction.csv')
```
